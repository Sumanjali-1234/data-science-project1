```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
g=sb.FacetGrid(df,col="sex",hue="smoker")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
#importing packages
import seaborn as sb
import matplotlib.pyplot as plt
#loading dataset using seaborn
df=sb.load_dataset('tips')
#df.head
#print(df.to_string())
sb.pairplot(df)
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,hue='size')
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
selected_vars=df[["total_bill","tip"]]
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,hue="size",palette="husl")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,diag_kind='kde')
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
sb.pairplot(df,kind='reg')
plt.show()
```

```python
#controlling the markers
import seaborn as sns
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
sns.pairplot(df,hue='sex',markers=["o","s"])
plt.show()
```

```python
g=sb.pairplot(df,hue='day')
g.fig.suptitle("pairplot of Tips Dataset",y=1.02)
g.set(xticks=[])
plt.show()
```
